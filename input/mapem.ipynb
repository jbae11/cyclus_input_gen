```python
import sqlite3 as lite
import folium
```

```python
m = folium.Map()
```

```python
con = lite.connect('./cyclus.sqlite')
con.row_factory = lite.Row
cur = con.cursor()
q = cur.execute('SELECT * FROM AgentPosition').fetchall()
pos_dict = {r['Prototype']: (r['Latitude'], r['Longitude']) for r in q if 'Reactor' in r['Spec']}
```

```python
for reactor, loc in pos_dict.items():
    folium.Marker(location=loc, popup=reactor).add_to(m)
```
